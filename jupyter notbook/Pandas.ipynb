Handle Missing Data

```python
import pandas as pd
df = pd.read_csv("weather_data (1).csv", parse_dates=["day"])
type(df.day[0])
df.set_index("day", inplace=True)
df
```

```python
new_df = df.fillna(0) # Fills the NaN values with 0
new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

```python
new_df = df.fillna(method="ffill") # Forward fill fills the NaN values with the previous value
new_df
```

```python
new_df = df.fillna(method="bfill") # Backward fill fills the NaN values with the next value
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```

Mearge the data

```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando", "baltimore"],
    "temperature": [21, 14, 35, 32],
})

ax = df1.plot.pie(y="temperature", labels=df1["city"], autopct="%1.1f%%", shadow=True, explode=(0, 0, 0.1, 0), startangle=90)
ax.set_ylabel('')
ax.set_title('City Temperature Distribution')

plt.show()
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san francisco"],
    "humidity": [65,68,71],
})

ax2 = df2.plot.pie(y="humidity", labels=df2["city"], autopct="%1.1f%%", shadow = True, explode = (0, 0.1, 0))
ax2.set_ylabel('')
ax2.set_title('City Humidity Distribution')
plt.show()


```

```python
df = pd.merge(df1,df2,on = "city", how = "outer",indicator=True)
df
```

```python
df1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando", "baltimore"],
    "temperature": [21, 14, 35, 38],
    "humidity": [60, 80, 40, 50],
})
df1
```
